```python
import pandas as pd
df = pd.read_csv('/content/customer_data_fact_202301221401.csv')
```

```python
df.head(5)
```

```python
df2 = df.drop(columns=['CLIENTNUM'])
df2.head(5)
```

```python
#https://medium.easyread.co/membuat-proyek-machine-learning-dengan-python-part-1-8e8a03095636
from pandas.io.formats.info import DataFrameTableBuilderNonVerbose
# Import LabelEncoder dari module sklearn
from sklearn.preprocessing import LabelEncoder

# Menyalin / copy dataframe agar dataframe awal tetap utuh
dataframe_int = df2.copy()

# Membuat objek/instance yang bernama encoder
encoder = LabelEncoder()

# Membuat list dari nama kolom data kategori
categorical_data = ['status','Gender','Education_Level','Marital_Status','Income_Category','Card_Category']

# Mengubah setiap data kategori menjadi numerik dengan encoder
for kolom in categorical_data:
    dataframe_int[kolom] = encoder.fit_transform(df2[kolom])
	
# Sekarang data sudah berupa angka sepenuhnya
dataframe_int.head()
```
